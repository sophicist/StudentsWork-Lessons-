```python
import pandas as pd
import os
os.listdir()
```

```python
dt = pd.read_excel('weights.xlsx')
dt.head(10)
```

```python
# calculate the  outlier given the data
LL = 8.5
UL = 112.5
new= []
for i in dt['Weight']:
    if (i<LL)|(i>UL):
        new.append("outlier")
    else:
        new.append('Normal')
dt["John_outlier"] = new
dt.head(10)
```
