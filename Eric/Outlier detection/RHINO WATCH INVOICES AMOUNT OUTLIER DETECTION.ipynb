```python
df = pd.ExcelFile("INVOICES 1.xlsx")
print(df.sheet_names)
```

```python
dt = pd.read_excel(df,' anlysis_DS',
                  skiprows = 2
                  )
dt.iloc[5:,:]
```

```python

dt['DATE'] = [str(i).replace("//","/") for i in dt['DATE']]
dt['DATE'] = pd.DatetimeIndex(dt['DATE'])
dt['INV. AMOUNT(EURO)'] = [float(str(i).replace('EUROS','0')) for i in dt['INV. AMOUNT(EURO)']]
dt['INV. AMOUNT(USD)'] = [float(str(i).replace('EUROS','0')) for i in dt['INV. AMOUNT(USD)']]
dt['SAFARIS'] = [float(str(i).replace(' ','0')) for i in dt['SAFARIS']]
```

```python
dt.info()
```

```python
month = []
for date in dt['DATE']:
    mon = date.month_name()
    year  = date.year
    month.append(mon)

dt['month'] = month
```

```python
dt.columns = [i.strip() for i in dt.columns]
```

```python
dy = dt[['DATE','CLIENT NAME','INV. AMOUNT(USD)','ACCOMODATION','SAFARIS','LOCAL AIR TRAVEL CHARGES','month']]
dy['year'] = [i.year for i in dy['DATE']]
dy.head()
```

```python
# how much do the company make  each year
dy[['ACCOMODATION','SAFARIS','LOCAL AIR TRAVEL CHARGES']].sum()
```
