```python
from csv import reader

csv_file = "Convertible_1.csv"
def csreader(csv_file):
    data = []
    with open(csv_file,"r") as f:
        csv_reader = reader(x.replace('\0', '') for x in f)
        # Iterate over each row in the csv using reader object
        for row in csv_reader:
            try:
                # row variable is a list that represents a row in csv
                data.append(row)
            except:
                pass
    x = data[0]
    y = data[1:]
    df = pd.DataFrame(columns = x)
    d = 0
    for i in y:
        df.loc[d] = i
        d+=1
    print(f"The read csv has shape:{d} rows and {len(x)} columns")
    return df
```
