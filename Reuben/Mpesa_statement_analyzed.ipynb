```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
details = data['Details'].unique()
```

```python
# Sample transactions data
transactions = ["Funds received from John", "Payment sent to Sarah", "Funds received from Mark", "Payment sent to Lisa", "Funds received from David"]

money_received = []
money_sent = []

for transaction in transactions:
    if "Funds received from" in transaction:
        money_received.append(transaction)
    elif "Payment sent to" in transaction:
        money_sent.append(transaction)

print("Money Received:", money_received)
print("Money Sent:", money_sent)
```

```python
money_received = 4000
money_sent = 2000

# Pie chart
labels = ['Money Received', 'Money Sent']
sizes = [money_received, money_sent]
colors = ['green', 'red']
explode = (0.1, 0)  # explode the Money Received slice

plt.pie(sizes, labels=labels, colors=colors, startangle=90)
plt.axis('equal')
plt.title('Money Flow')
plt.show()

```
